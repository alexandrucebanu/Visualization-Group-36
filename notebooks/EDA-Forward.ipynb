```python
##  Imports  ##
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
## Config
DATADIR="data/"
```

```python
## Loading the data files

areas  = ["shooting","possession","playingtime","passing","misc"];
frames = []

counter=0
for area in areas:
    fileName = "player_"+area+".csv"
    frame    =  pd.read_csv(DATADIR+fileName)
    frames.append(frame)
```

```python
## Merging all frames into a single one

df=frames[0]
for i in range(1,len(frames)):
    df=pd.merge(df,frames[i])
```

```python
## Filtering the players to only get the ones who play in the forward position
df=df[df['position']=="FW"]
```

```python
df
```

```python
def getAgeYears(ageString):
    return int(ageString.split('-')[0])
df['age']=(df['age']).map(getAgeYears)
```
